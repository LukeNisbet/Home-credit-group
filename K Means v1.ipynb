```python
import numpy as np
import pandas as pd
import pylab as pl

import matplotlib.pyplot as plt

import sklearn
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler as scale

import sklearn.metrics as sm
from sklearn.metrics import confusion_matrix, classification_report
```

```python
apps  = pd.read_pickle('C:\\Users\\luken\\Documents\\Home Credit Group\\Data\\train_only_app_features.pkl')
apps.sample(10)
```

# One Hot Encoding
Transform catagorical to binary dummies

```python
# Encode the catgorical varibles with one hot 
def one_hot_encode_it(dset):

    print('The following varibles have been one hot encoded.')
    for column in dset.columns:
        if dset[column].dtype == type(object):
            x = pd.get_dummies(dset[column],prefix=column)
            dset = dset.join(x,on='SK_ID_CURR').drop(str(column),1)
            del x
            print(column)
    return dset

apps = one_hot_encode_it(apps)
apps.sample(10)
```

# Scale data
Scale the input data to avoid large value dominating the euclidian space calculation


```python
## Baisc data check to confirm that data has no NAN (missings) or infinity values
# np.any(np.isnan(apps))
np.any(np.isfinite(apps))
# apps = apps.fillna(0)
# 
# print(apps.loc[[445398]])


scaled_app = scale(apps)
# NOTE: Scaled_app is not longer a pandas dataframe, it is not a numpy array
scaled_app[:10]
```

# Builing and running the model

```python

Nc = range(2, 30)

kmeans = [KMeans(n_clusters=i) for i in Nc]

kmeans

score = [kmeans[i].fit(scaled_app).score(scaled_app) for i in range(len(kmeans))]

print(list(Nc))

print(score)

plt.plot(list(Nc),score)

plt.xticks(list(Nc))

print(plt.xticks())

plt.xlabel('Number of Clusters')

plt.ylabel('Score')

plt.title('Elbow Curve')

plt.show()
```

```python

targets = apps['TARGET']
scores = np.empty(1)

for i in range(2,30):
    col_name = str(i)+' clusters'
    print(col_name)
    kmeans = KMeans(n_clusters=i)
    temp = kmeans.fit(scaled_app)
    temp_score = temp.score(scaled_app)
    temp_label = temp.labels_
    labels = pd.DataFrame(data = temp_label
                          ,columns = [str(i)+' clusters'])
    targets = pd.concat([targets,labels], axis=1)
    scores = np.append(scores,temp_score)
    
    print(scores)
    print(targets.sample(10))
    
```
